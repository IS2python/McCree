```python
import csv
import math as m
f = open('f.csv')
c = csv.reader(f)

def standardG(g, o):
    sin2 = (m.sin(m.radians(o)))**2
    return g*(1+0.005278895*sin2+0.000023462*sin2**2)

def freeAirCorrection(h):
    return 0.308*h

def freeAirAnomaly(g, h, o):
    return (g - standardG(g, o)+freeAirCorrection(h))

def bouguerCorrection(p, h):
    return 0.0419 * p * h

def bouguerAnomaly(g, h, o, p):
    return freeAirAnomaly(g, h, o) - bouguerCorrection(p, h)

def terrainCorrection(g, h, o, p, gBC):
    return gBC - bouguerAnomaly(g,h,o,p)

for i in c:
    l = i
    location = str(l[0])
    latitude = float(l[1])
    longitude = float(l[2])
    h = float(l[3])
    g = float(l[4])
    gBC = float(l[5])
    stone = str(l[6])
    p = float(l[7])
    dgt = terrainCorrection(g, h, latitude, p, gBC)
    print(dgt)
location = input()
for i in c:
    if location == str(i[0]):
        x = float(i[1])
        y = float(i[2])
xx = 0
yy = 0
v = 100000
for i in c:
    location = str(i[0])
    latitude = float(i[1])
    longitude = float(i[2])
    vv = (x-latitude)**2 + (y-longitude)**2
    if x != latitude and vv < v :
        xx = latitude
        yy = longitude
for i in c:
    location = str(i[0])
    latitude = float(i[1])
    longitude = float(i[2])
    if xx == latitude and yy == longitude:
        print(location)
f.close()
```
